```python
import pandas as pd
import re

# ==============================
# 1) إعداد المسارات
# ==============================
input_file = "التسكين الوضع الحالي 2026.xlsx"
output_file = "department_role_employee_count_cleaned.xlsx"

# ==============================
# 2) قراءة الملف
# ==============================
df = pd.read_excel(input_file)

# ==============================
# 3) تنظيف أسماء الأعمدة
# ==============================
df.columns = [str(col).strip() for col in df.columns]

# ==============================
# 4) توحيد أسماء الأعمدة
# ==============================
rename_map = {
    'المسمي الوظيفي': 'المسمى الوظيفي',
    'المسمي الوظيفي ': 'المسمى الوظيفي',
    'المسمى الوظيفي ': 'المسمى الوظيفي',
    'الإدارة- code': 'كود الإدارة',
    'الإدارة - code': 'كود الإدارة',
    'الإدارة -code': 'كود الإدارة',
    'الإدارة-code': 'كود الإدارة',
}

df = df.rename(columns=rename_map)

# ==============================
# 5) التأكد من الأعمدة المطلوبة
# ==============================
required_columns = ['الإدارة', 'كود الإدارة', 'المسمى الوظيفي']
missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"الأعمدة التالية غير موجودة في الملف: {missing_columns}")

# ==============================
# 6) اختيار الأعمدة المطلوبة فقط
# ==============================
df = df[required_columns].copy()

# ==============================
# 7) تنظيف القيم النصية
# ==============================
def clean_arabic_text(value):
    if pd.isna(value):
        return pd.NA

    value = str(value).strip()

    # إزالة المسافات الزائدة داخل النص
    value = re.sub(r'\s+', ' ', value)

    # توحيد بعض الأحرف العربية
    value = re.sub(r'[إأآا]', 'ا', value)
    value = value.replace('ى', 'ي')
    value = value.replace('ؤ', 'و')
    value = value.replace('ئ', 'ي')

    # إزالة أي فراغات إضافية مرة ثانية
    value = re.sub(r'\s+', ' ', value).strip()

    # اعتبار النصوص الفارغة كقيم مفقودة
    if value == "" or value.lower() == "nan":
        return pd.NA

    return value

df['الإدارة'] = df['الإدارة'].apply(clean_arabic_text)
df['المسمى الوظيفي'] = df['المسمى الوظيفي'].apply(clean_arabic_text)

# كود الإدارة نخليه كنص للحفاظ على التنسيق
df['كود الإدارة'] = df['كود الإدارة'].astype(str).str.strip()
df['كود الإدارة'] = df['كود الإدارة'].replace(['', 'nan', 'None'], pd.NA)

# ==============================
# 8) حذف الصفوف التي فيها قيم ناقصة
# ==============================
df = df.dropna(subset=['الإدارة', 'كود الإدارة', 'المسمى الوظيفي']).copy()

# ==============================
# 9) حساب عدد الموظفين لكل مسمى في كل إدارة
# ==============================
result = (
    df.groupby(['الإدارة', 'كود الإدارة', 'المسمى الوظيفي'], dropna=False)
      .size()
      .reset_index(name='عدد الموظفين')
)

# ==============================
# 10) ترتيب النتيجة
# ==============================
result = result.sort_values(
    by=['الإدارة', 'كود الإدارة', 'عدد الموظفين', 'المسمى الوظيفي'],
    ascending=[True, True, False, True]
).reset_index(drop=True)

# ==============================
# 11) حفظ النتيجة في ملف Excel
# ==============================
result.to_excel(output_file, index=False)

print(f"تم حفظ الملف بنجاح: {output_file}")
print(f"عدد الصفوف في النتيجة: {len(result)}")
```
